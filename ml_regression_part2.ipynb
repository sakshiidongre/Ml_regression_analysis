```python
import pandas as pd
import matplotlib.pyplot as plt
mt = pd.read_csv('mtcars.csv')

sm = pd.DataFrame(mt)
sm.head()
```

```python
sm.corr()
```

```python
from sklearn import linear_model

model = linear_model.LinearRegression()
x = sm[['wt']]
y = sm[['mpg']]
model.fit(x,y)
```

```python
ypred =model.predict(x)
plt.scatter(sm['wt'],sm['mpg'], c='r')
plt.plot(sm['wt'],ypred,'kD:')
```

```python
model.score(x,y)
```

```python
X3 = sm[['cyl',	'disp','hp','drat','wt','vs','gear','carb']]
model3 = linear_model.LinearRegression()
model3.fit(X3,y)
model3.predict([[6,160,90,2.7,2.5,1,4,2]])
```

```python
model3.score(X3,y)
```
